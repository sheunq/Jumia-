```python
import pandas as pd
from bs4 import  BeautifulSoup as bs
import requests
import re
```

```python
def data_scraping(category,type):
  z=[]
  x=[]
  for i in range(1,51):
    url=f'https://www.jumia.com.ng/{category}/?page={i}#catalog-listing'
    result=requests.get(url).content
    jum=bs(result,'lxml')
    item_name=jum.find_all(f'{type}',class_='name')
    item_price=jum.find_all('div',class_='prc')
    for c in item_name:
      z.append(c.text)
    for d in item_price:
      x.append(d.text)
  return z,x    
```

## Health and Beauty

```python
health_item,health_item_price=data_scraping('health-beauty','h3')
```

```python
health=pd.DataFrame({'Item':health_item,'category':'health','Price':health_item_price[:len(health_item)]})
health
```

```python
jumia_items.drop(['p','c'],inplace=True,axis=1)
```
